```python
# Import libaries
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
import pandas as pd
from datetime import datetime
pd.options.display.max_colwidth = 400
```

```python
url = "https://www.rev.com/blog/transcripts/crypto-ceos-testify-before-house-financial-services-hearing-transcript"
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
```

```python
soup
```

```python
result = soup.body.find_all("p")
result
```

```python
transcript = []
for tran in result:
    transcript1 = {}
    transcript1["conversation"] = tran.text
    transcript.append(transcript1)
```

```python
transcript_dic = pd.DataFrame(transcript)
```

```python
transcript_dic.head(10)
```

```python
transcript_dic.tail(10)
```

```python
transcript_dic.drop([0,1,2,3,4, 650,651,652,653,654], inplace=True)
```

```python
transcript_dic1 = transcript_dic["conversation"].str.split(":", expand=True, n=1)
```

```python
transcript_dic1.columns = ["speaker", "speech"]
```

```python
transcript_dic1
```

```python
transcript_dic2 = transcript_dic1["speech"].str.split("\n", expand=True, n=1)
```

```python
transcript_dic2
```

```python
transcript_dic2= transcript_dic2.join(transcript_dic1 )
```

```python
transcript_dic2.drop("speech", axis=1, inplace=True)
```

```python
transcript_dic2.columns = ["time", "speech", "speaker"]
```

```python
transcript_dic2 = transcript_dic2[["time", "speaker", "speech"]]
```

```python
transcript_dic2["time"] = transcript_dic2["time"].map(lambda x: x.replace("(", "").replace(")", "").replace(" ",""))
```

```python
transcript_dic2["time"] = transcript_dic2["time"].map(lambda x: "00:{}".format(x) if len(x)<7 else x)
```

```python
transcript_dic2.info()
```

```python
transcript_dic2["time"] = pd.to_datetime(transcript_dic2["time"])
```
